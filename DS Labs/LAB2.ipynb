Mean for Numeric

```python
import pandas as pd
import numpy as np

#Sample dataset
df = pd.DataFrame({
    'Age': [25,30,np.nan,40,35],
    'Department':['HR', 'Finance','Finance',np.nan,'IT']})
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Department'].fillna(df['Department'].mode()[0], inplace=True)
print(df)
```

Forward Fill

```python
import pandas as pd
import numpy as np

#Sample dataset
df = pd.DataFrame({
    'Age': [25,30,np.nan,40,35],
    'Department':['HR', 'Finance','Finance',np.nan,'IT']})
df_ffill = df.copy()
df_ffill.fillna(method='ffill', inplace=True)
print(df_ffill)
```

Backward Fill

```python
import numpy as nd
import pandas as pd
#Sample dataset
df = pd.DataFrame({
    'Age': [25,30,np.nan,40,35],
    'Department':['HR', 'Finance','Finance',np.nan,'IT']})
df_bfill = df.copy()
df_bfill.fillna(method='bfill', inplace=True)
print(df_bfill)
```

Deletion

```python
# Drop rows with missing values
import pandas as pd
import numpy as np
# Sample dataset
df = pd.DataFrame({  'Age': [25, 30, np.nan, 40, 35],
'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']})

df_drop_rows = df.dropna()
print("After dropping rows:\n", df_drop_rows)


df_drop_cols = df.dropna(axis=1)
print("After dropping columns:\n", df_drop_cols)
```

Removing duplicates

```python
#1. Exact Match Removal

import pandas as pd

#sample dataset
df = pd.DataFrame({
    'ID':[1,2,2,3,4,4],
    'Name':['Alice','Bob','Bob','Charlie','Dave','Dave'],
    'Age':[25,30,30,35,40,40],

})
print("Original Data:\n",df)

#remove exact duplicates
df_exact=df.drop_duplicates()
print("\n After Exact Match Removal:\n",df_exact)
```

```python
#2. Subset based Removal

df_subset_id = df.drop_duplicates(subset=['ID'])
print("\n After Subset based Removal:\n",df_subset_id)

#removing duplicates based on name
df_subset_name = df.drop_duplicates(subset=['Name'])
print("\n After Subset based Removal:\n",df_subset_name)

```

Data Standardization

```python
import pandas as pd

# Sample dataset with inconsistent date formats
df = pd.DataFrame({
    'Date': ['2025-01-05', '05/01/2025', 'Jan 5, 2025', '2025.01.05']
})

# Convert all dates to standard ISO format (YYYY-MM-DD)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%Y-%m-%d')

print(df)
```

Case Normalization

```python
# Sample dataset with inconsistent text cases
df = pd.DataFrame({
    'Name': ['Alice', 'BOB', 'charlie', 'DAVID']
})

# Convert all names to lowercase
df['Name_lower'] = df['Name'].str.lower()

# Convert all names to uppercase
df['Name_upper'] = df['Name'].str.upper()

print(df)
```
