Week 3: Implementation of Similarity and Dissimilarity Measures for Data Analysis

1. Vector-Based Distance Measures: Used mainly for numerical data


```python
# 1.Using Euclidean Distance

import numpy as np
from scipy.spatial import distance

A= np.array([2,4,6])
B = np.array([5,1,9])

#Euclidean Distance
euclidean_distance = distance.euclidean(A,B)
print("Euclidean Distance:",euclidean_distance)

#Similarity
sim = 1/(1+euclidean_distance)
print("Similarity:",sim)
```

```python
# 2.Using Manhattan Distance

import numpy as np
from scipy.spatial import distance

manhattan_dist = distance.cityblock(A,B)
print("Manhattan Distance:",manhattan_dist)

#Similarity
sim = 1/(1+manhattan_dist)
print("Similarity:",sim)
```

```python
# 2.Using Minkowski  Distance

# It is a general form, it can become manhattan when p=1 and euclidean when p=2,
# so that is why we taken p=3 otherthan 1 and 2

import numpy as np
from scipy.spatial import distance

minkowski_dist = distance.minkowski(A,B,p=3)
print("Minkowski Distance(p=3):",minkowski_dist)

#Similarity
sim = 1/(1+minkowski_dist)
print("Similarity:",sim)
```

2. Statistical Measures

Statistical measures are used to find the relationship between variables

```python
# Pearson Correlation

import pandas as pd
from scipy.stats import pearsonr

tips = pd.read_csv("tips.csv")

corr,p_value = pearsonr(tips["total_bill"],tips["tip"])
print("Correlation:",corr)
print("P-value:",p_value)

# Spearman Rank correlation
# Spearman Rank correlation which we use when data is not perfectly linear then
# we use these kind of ranking methods

from scipy.stats import spearmanr

df = pd.DataFrame({
    'X':[10,20,30,40,50],
    'Y':[12,18,33,47,55]
})

corr_value , p_value = spearmanr(df["X"],df["Y"])
print("Correlation:",corr_value)
print("P-value:",p_value)

#df = pd.read_csv("tips.csv")
print(df.corr(method='spearman'))
```

3. Edit-Based Measures

```python
def hamming_distance(str1,str2):
  if len(str1) != len(str2):
    raise ValueError("Strings must be of equal length")
  return sum(ch1 != ch2 for ch1, ch2 in zip(str1, str2))

s1 = "karolin"
s2= "kathrin"
distance = hamming_distance(s1,s2)
print(f"Hamming Distance between'{s1}' and '{s2}': {distance}")
```

4. Token-Based Measures

```python
# 1. Jaccard Index

def jaccard_index(str1,str2):
  set1, set2 = set(str1.split()), set(str2.split())
  intersection = set1.intersection(set2)
  union = set1.union(set2)
  return len(intersection) / len(union)

s1 = "data science is fun"
s2 = "science makes data useful"

print("Jaccard Index:", jaccard_index(s1,s2))
```

```python
# 2. Cosine Similarity

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

s1 = "data science is fun"
s2 = "science makes data useful"

vectorizer = CountVectorizer().fit([s1,s2])
vectors = vectorizer.transform([s1,s2])
cos_sim = cosine_similarity(vectors[0], vectors[1][0][0])
print("Cosine Similarity:", cos_sim)
```

5. Sequence-Based Measures

```python
#lonest common subsequence
def lcs_length(X, Y):
  m, n = len(X), len(Y)

  # Create matrix
  dp = [[0] * (n + 1) for _ in range(m + 1)]

  # Build matrix
  for i in range(m):
    for j in range(n):

      if X[i] == Y[j]:
        dp[i + 1][j + 1] = dp[i][j] + 1

      else:
        dp[i + 1][j + 1] = max(dp[i][j + 1], dp[i + 1][j])

  return dp[m][n]


# Example usage
seq1 = "ABCDEF"
seq2 = "AEBDF"

length = lcs_length(seq1, seq2)

print("Longest Common Subsequence Length:", length)
```
